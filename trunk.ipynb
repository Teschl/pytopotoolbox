```python
import topotoolbox as tt
import numpy as np
import matplotlib.pyplot as plt 
```

```python
dem = tt.load_dem('perfectworld')
flow = tt.FlowObject(dem)
stream = tt.StreamObject(flow,threshold=1000,units='pixels')
```

```python
plt.subplot()
plt.imshow(dem, cmap='terrain')
stream.plot(color='b')
stream.trunk().plot(color='r')
plt.show()
```

```python
acc = flow.flow_accumulation()
down = stream.ezgetnal(acc)

temp = stream.trunk(flow_accumulation=acc)
plt.imshow(dem, cmap='terrain')
temp.plot(color='r')
plt.show()
temp = stream.trunk(downstream_distance=down)
plt.imshow(dem, cmap='terrain')
temp.plot(color='r')
plt.show()
```
